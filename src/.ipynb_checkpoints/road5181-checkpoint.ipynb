```python
import pandas as pd
import geopandas as gpd
```

4. 사고포인트들의 도로 파악 분석 파트(사고난 도로의 도로명, 도로폭, 교통량 등 분석 필요)

```python
road5181 = gpd.read_file('../data/road/Z_KAIS_TL_SPRD_MANAGE_11000.shp', encoding='euc-kr')
```

```python
# buffer
buff_road = road5181.buffer(road5181['ROAD_BT'], cap_style=2)
# buffer된 폴리곤을 기존 linestring 속성과 대체
road5181.geometry = buff_road
road5181.plot()
```

```python
road5181.to_file('../data/road/buffer_road_5181.shp', encoding='euc-kr')
```
