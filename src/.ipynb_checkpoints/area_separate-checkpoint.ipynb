```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import geoplot
import rasterio
from rasterio.plot import show
import geopandas as gpd
from ipyleaflet import *
import seaborn as sns
%matplotlib inline
import fiona
from matplotlib import pyplot as plt
from rasterio.plot import show
from rasterio.mask import mask
from ipywidgets import IntSlider, ColorPicker, jslink
from ipyleaflet import Map, basemaps, basemap_to_tiles, DrawControl
```

```python
seoul_gu = gpd.read_file('../data//UPIS_C_UQ111_ORG.shp', encoding='euc-kr').to_crs(5181)
seoul_gu.plot()
```

```python
pm_accident_csv = pd.read_csv('../data/PM/pm4326.csv')
pm_geometry = gpd.points_from_xy(x = pm_accident_csv[''], y = pm_accident_csv[''])
pm_geometry = gpd.GeoDataFrame(pm_accident_csv, geometry = pm_geometry)
pm_geometry.crs = 'epsg:4326'
pm_geometry.to_crs(5181, inplace=True)
pm_geometry.rename(columns = {'':'longitude','':'latitude'}, inplace = True)
```

```python
base = seoul_gu.plot(color='#cccccc', edgecolor="white")
ax = pm_geometry.plot(ax=base, marker='o', color='red', markersize=10)
ax.set_axis_off()
plt.show()
```
