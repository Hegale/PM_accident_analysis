```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

4. 사고포인트들의 도로 파악 분석 파트(사고난 도로의 도로명, 도로폭, 교통량 등 분석 필요)

```python
road5174 = gpd.read_file('../data/road/Z_KAIS_TL_SPRD_MANAGE_11000.shp', encoding='euc-kr')
road5174.crs = 'epsg:5174'
road4326 = road5174.to_crs(4326)
```

```python
road4326['도로폭'] = road4326['ROAD_BT'] * (0.00001 / 1.1)
road4326['도로폭']
```

```python
# buffer
buff_road = road4326.buffer(road4326['도로폭'], cap_style=2)
# buffer된 폴리곤을 기존 linestring 속성과 대체
road4326.geometry = buff_road
road4326.plot()
```

```python
# 총 교통량 데이터 불러오기 : traffic
traffic = pd.read_csv('../data/traffic/merged_traffic.csv', encoding='euc-kr')
```

```python
# 교통량 데이터의 괄호 안 값을 제거하고 교통량 통합하기
pattern = r'\([^)]*\)'
traffic['지점명'] = traffic['지점명'].str.replace(pat = pattern, repl = r'', regex=True)
traffic_merged = traffic.groupby('지점명').sum()
traffic_merged['RN_simple'] = traffic_merged.index
```

```python
# 교통량과의 매칭을 위해 'n길' 삭제
pattern = r'[0-9].*$'
road4326['RN_simple'] = road4326['RN'].str.replace(pat = pattern, repl = r'', regex=True)
road_lt_sum = road4326.groupby('RN_simple').agg({
    'ROAD_LT' : np.sum})
```

```python
# 교통량을 길이로 나눈 속성을 road4326에 추가
traffic_length = pd.merge(road_lt_sum, traffic_merged, how='left', on='RN_simple')
traffic_length['traffic'] = traffic_length['총 교통량'] / traffic_length['ROAD_LT']
traffic_length.to_csv('../data/traffic/traffic_by_length.csv', encoding='euc-kr')
road4326 = pd.merge(road4326, traffic_length, how='left', on = 'RN_simple')
```

```python
road4326.to_file('../data/road/buffer_road_4326.shp', encoding='euc-kr')
```
