```python
import pandas as pd
import geopandas as gpd
import re
import numpy as np
import geoplot
from ipyleaflet import Map, Marker, GeoData
import matplotlib.pyplot as plt
from shapely.ops import unary_union
```

### 사고포인트들의 도로 파악 분석 파트(사고난 도로의 도로명, 도로폭, 교통량 등 분석 필요)

1. 필요한 데이터(사고 지점 points, 실폭도로 polygon, 교통량 csv) 불러오기
2. 실폭도로 데이터와 교통량을 합치기 위해 교통량 데이터의 괄호() 내용 없애고, 중복 데이터는 합치기
3. 실폭도로 데이터의 length를 도로명별로 합치고(폴리곤을 합치는게 아님), 그걸 교통량 데이터의 도로명과 매치하여 교통량/구간길이를 만들기
4. 3을 다시 실폭도로 데이터와 일괄 매칭.....

```python
# 사고 지점 데이터 불러오기 : pm_4326
pm_accident_csv = pd.read_csv('../data/PM/pm4326.csv')
pm4326 = gpd.points_from_xy(x = pm_accident_csv['경도'], y = pm_accident_csv['위도'])
pm4326 = gpd.GeoDataFrame(pm_accident_csv, geometry = pm4326)
pm4326.crs = 'epsg:4326'
```

```python
# 실폭도로 polygon 데이터 불러오기 : road4326
road4326 = gpd.read_file('../data/road/buffer_road_4326.shp', encoding='euc-kr')
```

```python
# 총 교통량 데이터 불러오기 : traffic
traffic = pd.read_csv('../data/road/총 교통량_지점별.csv', encoding='euc-kr')
```

```python
# 교통량 데이터의 괄호 안 값을 제거하고 교통량 통합하기
pattern = r'\([^)]*\)'
traffic['지점명'] = traffic['지점명'].str.replace(pat = pattern, repl = r'', regex=True)
traffic_merged = traffic.groupby('지점명').sum()
traffic_merged['RN_simple'] = traffic_merged.index
```

```python
# 교통량과의 매칭을 위해 'n길' 삭제
pattern = r'[0-9].*$'
road4326['RN_simple'] = road4326['RN'].str.replace(pat = pattern, repl = r'', regex=True)
road_lt_sum = road4326.groupby('RN_simple').agg({
    'ROAD_LT' : np.sum})
```

```python
# 교통량을 길이로 나눈 속성을 road4326에 추가
traffic_length = pd.merge(road_lt_sum, traffic_merged, how='left', on='RN_simple')
traffic_length['traffic'] = traffic_length['총 교통량'] / traffic_length['ROAD_LT']
traffic_length.to_csv('../data/traffic/traffic_by_length.csv', encoding='euc-kr')
road4326 = pd.merge(road4326, traffic_length, how='left', on = 'RN_simple')
```

```python
# ipyleaflet 으로 지도 불러오기
center = (37.5642153, 127.0016985)
seoul = Map(center=center, zoom=11)

# 도로 올리기
road_polygon = GeoData(geo_dataframe = road4326,
    style={'color': 'black', 'fillColor': '#fa8072', 'opacity':0.05, 'weight':1.9, 'dashArray':'2', 'fillOpacity':0.6},
    hover_style={'fillColor': 'red' , 'fillOpacity': 0.2},
    name = 'Road')

# pm 사고지점을 geodata로
pm_points = GeoData(geo_dataframe = pm4326,
    style={'color': 'black', 'radius':8, 'fillColor': '#3366cc', 'opacity':0.5, 'weight':1.9, 'dashArray':'2', 'fillOpacity':0.6},
    hover_style={'fillColor': 'red' , 'fillOpacity': 0.2},
    point_style={'radius': 5, 'color': 'red', 'fillOpacity': 0.8, 'fillColor': 'blue', 'weight': 3},
    name = 'Release')

seoul.add_layer(road_polygon)
seoul.add_layer(pm_points);
```

```python
# 사고지점을 도로 데이터에 inner 조인
merged_pm_inner = gpd.sjoin(pm4326, road4326, how='inner', op='intersects')
merged_pm_inner
```

```python
# pm 사고지점을 도로 데이터에 outer left 조인
merged_pm = gpd.sjoin(pm4326, road4326, how='left', op='intersects')
merged_pm.head()
```

```python
# 폴리곤이 겹친 문제로 같은 위치에 중복 생성된 점 삭제.
# 이후, 대략적인 개수 확인을 위해 점 도시
merged_pm.drop_duplicates(['geometry'], inplace = True)
merged_pm.plot()
```

```python
# 교통량 도시
plt.title('traffic')
plt.hist(merged_pm['traffic'], color = 'r', alpha = 0.5)
```

```python
# 교통량 도시
plt.title('road_bt')
plt.hist(merged_pm['ROAD_BT'], bins = 7, color = 'g', alpha = 0.5)
```
