```python
import numpy as np  # gridded data
import os  # file managment
import csv  # for saving results in csv format
import geopandas as gpd
import pandas as pd
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
```

https://gis.stackexchange.com/questions/317391/python-extract-raster-values-at-point-locations

```python
result_raster = rasterio.open('../data/raster/result_raster.tif')
pm_accident_csv = pd.read_csv('../data/PM/pm4326.csv')
pm_points = gpd.points_from_xy(x = pm_accident_csv['경도'], y = pm_accident_csv['위도'])
pm_points = gpd.GeoDataFrame(pm_accident_csv, geometry = pm_points)
pm_points.crs = 'epsg:4326'
```

```python
# Read points from shapefile
pts = pm_points
pts = pts[['geometry']]
pts.index = range(len(pts))
coords = [(x,y) for x, y in zip(pts.geometry.x, pts.geometry.y)]

# Open the raster and store metadata
src = rasterio.open('../data/raster/result_raster.tif')

# Sample the raster at every point location and store values in DataFrame
pts['Raster Value'] = [x for x in src.sample(coords)]
```

```python
for i in range(len(pts)):
    pts.iloc[i, 1] = pts.iloc[i, 1][0]
```

```python
# 경사도 도시
plt.title('raster value')
plt.hist(pts['Raster Value'], color = 'g', alpha = 0.5)
```

```python
# seoulslope위에 사고포인트(df) 결합
fig, ax = plt.subplots(1,1,figsize=(10,10))

# transform rasterio plot to real world coords
extent=[result_raster.bounds[0], result_raster.bounds[2], result_raster.bounds[1], result_raster.bounds[3]]
ax = rasterio.plot.show(result_raster, extent=extent, cmap='PuBuGn', ax=ax, title='results')

pm_points.plot(ax=ax, facecolor='red')
```
