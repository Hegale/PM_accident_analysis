### 데이터 불러오기

```python
road = gpd.read_file('../data/road/buffer_road_4326.shp', encoding='euc-kr')
```

```python
road.plot()
```

### 도로폭 래스터를 단순 래스터화
출처 : https://pygis.io/docs/e_raster_rasterize.html

```python
geom = [shapes for shapes in road.geometry]
raster = rasterio.open('../data/raster/point_to_raster.tif')
```

```python
# all touched = True
rasterized_T = features.rasterize(geom,
                                out_shape = raster.shape,
                                fill = 0,
                                out = None,
                                transform = raster.transform,
                                all_touched = True,
                                default_value = 1,
                                dtype = None)
```

```python
# all touched = False
rasterized_F = features.rasterize(geom,
                                out_shape = raster.shape,
                                fill = 0,
                                out = None,
                                transform = raster.transform,
                                all_touched = False,
                                default_value = 1,
                                dtype = None)
```

```python
type(rasterized_T)
# ndarray 자꾸 뒤집혀서나옴.. 얘네가 원점이 왼쪽 위라 그런가
```

```python
# plot raster, rasterized_F

fig, ax = plt.subplots(1, figsize = (10, 10))
show(rasterized_F[::-1], ax = ax)
plt.gca().invert_yaxis()
```

```python
# plot raster, rasterized_T

fig, ax = plt.subplots(1, figsize = (10, 10))
show(rasterized_T[::-1], ax = ax)
plt.gca().invert_yaxis()
```

### 도로 폭이 어떻게 분포되어 있나 확인

```python
# 도로폭 확인
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 도로폭 분포 도시
plt.title('ROAD_BT')
#interval = [5*i for i in range(10)]
#plt.xticks(interval)
plt.xlim(0, 60)
plt.hist(road['ROAD_BT'], bins = 100, color = 'r', alpha = 0.5)
```

### 도로폭 데이터에 가중치 부여

```python
road['weight'] = road['ROAD_BT'] / 10
```

### weight가 적용된 래스터 생성

```python
# create tuples of geometry, value pairs, where value is the attribute value you want to burn
geom_value = ((geom,value) for geom, value in zip(road.geometry, road['weight']))

# Rasterize vector using the shape and transform of the raster
road_raster = features.rasterize(geom_value,
                                out_shape = raster.shape,
                                transform = raster.transform,
                                all_touched = True,
                                fill = 0,   # background value
                                merge_alg = MergeAlg.replace,
                                dtype = np.float64)
```

```python
# Plot raster
fig, ax = plt.subplots(1, figsize = (10, 10))
show(road_raster[::-1], ax = ax)
plt.gca().invert_yaxis()
plt.savefig('../img/road_raster.png')
```

## 용도지역 래스터 만들기

```python
use_area = gpd.read_file('../data/use_area/use_area4326.shp', encoding='euc-kr').to_crs(4326)
geom = [shapes for shapes in use_area.geometry]
```

```python
# create tuples of geometry, value pairs, where value is the attribute value you want to burn
geom_value = ((geom,value) for geom, value in zip(use_area.geometry, use_area['weight']))

# Rasterize vector using the shape and transform of the raster
use_raster = features.rasterize(geom_value,
                                out_shape = raster.shape,
                                transform = raster.transform,
                                all_touched = True,
                                fill = 0,   # background value
                                merge_alg = MergeAlg.replace,
                                dtype = np.float64)
```

```python
# Plot raster
fig, ax = plt.subplots(1, figsize = (10, 10))
show(use_raster[::-1], ax = ax)
plt.gca().invert_yaxis()
plt.savefig('../img/use_raster.png')
```

### 결과 저장

```python
from rasterio.crs import CRS
from rasterio.transform import from_origin
```

```python
merged_array = np.reshape(merged, (1, 210, 280))
driver = "GTiff"
dim = merged_array.shape
print(dim)
height = dim[1]
width = dim[2]
count = 1
dtype = merged_array.dtype
print(dtype)
crs = CRS.from_epsg(4326)
```

```python
xmin, xmax = 126.75, 127.2
ymin, ymax = 37.4, 37.7
xsize = (xmax-xmin)/280
ysize = (ymax-ymin)/210
# top left x coordinates, y, resolution of xpixel size, resolution of ypixel size
transform = from_origin(xmin, ymax, xsize, ysize)
```

```python
with rasterio.open("../data/raster/result_raster.tif", "w",
                   driver = driver,
                   height = height,
                   width = width,
                   count = count,
                   dtype = dtype,
                   crs = crs,
                   transform = transform) as dst:
    dst.write(merged_array)
```

```python
# png로 저장
with rasterio.open("../data/raster/result_raster.png", "w",
                   driver = driver,
                   height = height,
                   width = width,
                   count = count,
                   dtype = dtype,
                   crs = crs,
                   transform = transform) as dst:
    dst.write(merged_array)
```
