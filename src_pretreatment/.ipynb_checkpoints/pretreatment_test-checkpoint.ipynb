```python
# 국토교통부 표준링크, 구간길이 포함됨
nodeLink = gpd.read_file('./data/NODELINKDATA/MOCT_LINK.shp', encoding='euc-kr')
```

```python
# 서울시 구단위 shp불러오기
seoul_gu = gpd.read_file('./data/shp/gu5181.shp', encoding='euc-kr').to_crs(5181)
seoul_gu.plot()
seoul_gu.head()
```

```python
# 전국구 NODELINK를 서울만 남기기
seoul_nodeLink = nodeLink[(nodeLink['LINK_ID'].str[:3] >= '100') &
                          (nodeLink['LINK_ID'].str[:3] <= '124')]
seoul_nodeLink = seoul_nodeLink.to_crs(5181)
```

```python
seoul_nodeLink.head()
```

```python
# 파일로 저장
seoul_nodeLink.to_file('./data/seoul/nodeLink_seoul.shp', driver = 'ESRI Shapefile', encoding='euc-kr')
```

```python
# 신호등 데이터의 x, y 좌표를 통해 포인트로 변환
Lights_df = pd.read_csv('./data/seoul/서울시 신호등 관련 정보.csv', encoding='euc-kr')
seoul_lights = gpd.GeoDataFrame(Lights_df, geometry=gpd.points_from_xy(Lights_df.X좌표, Lights_df.Y좌표))
```

```python
# 횡단보도 데이터의 wkt를 통해 GeoDataFrame으로 변환
crossWalk_df = pd.read_csv('./data/seoul/서울시 대로변 횡단보도 위치정보.csv', encoding='euc-kr')
crossWalk_series = gpd.GeoSeries.from_wkt(crossWalk_df['노드 WKT'])
seoul_crossWalk = gpd.GeoDataFrame(crossWalk_df, geometry = crossWalk_series)
seoul_crossWalk.crs='epsg:4326'
seoul_crossWalk.to_crs(5181)
```
