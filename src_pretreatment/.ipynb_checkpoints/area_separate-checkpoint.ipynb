```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
grade = gpd.read_file('../data/use_area/UPIS_C_UQ111_ORG.shp', encoding='euc-kr')
```

```python
grade.plot()
```

### 주거, 상업, 녹지지역으로 분류

```python
residential = grade[grade['DGM_NM'].str.contains('주거', na = False)]
commercial = grade[grade['DGM_NM'].str.contains('상업', na = False)]
green = grade[grade['DGM_NM'].str.contains('녹지', na = False)]
```

### 사고 포인트들의 분포 확인

```python
pm_geometry = gpd.read_file('../data/PM/accident_points_geo.shp', encoding='euc-kr')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
point_intersection1 = gpd.sjoin(pm_geometry, residential.to_crs(5181), how='inner', op='intersects')
point_intersection1.plot()

point_intersection2 = gpd.sjoin(pm_geometry, commercial.to_crs(5181), how='inner', op='intersects')
point_intersection2.plot()

point_intersection3 = gpd.sjoin(pm_geometry, green.to_crs(5181), how='inner', op='intersects')
point_intersection3.plot()
```

```python
data=[['residential',len(point_intersection1)],['commercial',len(point_intersection2)],['green',len(point_intersection3)]]
df=pd.DataFrame(data,columns=["factor","count"])
print(df)
```

```python
df.plot(x="factor", y="count", kind="bar", rot=0, width=0.2, color= '#FF6633', figsize=(6,5))

plt.title('Number of accidents by factor')
plt.xlabel('factor')
plt.ylabel('count')
plt.ylim(0,45)
plt.rcParams['font.size']=14
plt.savefig('../data/accident_use_area.png')
```

### 미리 용도지역에 가중치 부여

```python
grade['weight'] = 0
grade['use'] = ""
```

```python
grade.loc[grade['DGM_NM'].str.contains('주거', na = False), 'weight'] = 3
grade.loc[grade['DGM_NM'].str.contains('주거', na = False), 'use'] = "residential"
grade.loc[grade['DGM_NM'].str.contains('상업', na = False), 'use'] = "commercial"
grade.loc[grade['DGM_NM'].str.contains('녹지', na = False), 'use'] = "green"
```

```python
grade.to_crs(4326, inplace = True)
```

### 분류된 용도지역 확인

```python
%matplotlib inline
fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = (14,4)
residential.plot(color='pink', alpha = .6, ax = ax)
commercial.plot(color='blue', alpha = .6, ax = ax)
green.plot(color='yellowgreen', alpha = .6, ax = ax)
```

```python
residential.to_file('../data/use_area/residential_area.shp', encoding='euc-kr')
commercial.to_file('../data/use_area/commercial_area.shp', encoding='euc-kr')
green.to_file('../data/use_area/green_area.shp', encoding='euc-kr')
grade.to_file('../data/use_area/use_area4326.shp', encoding='euc-kr')
```
